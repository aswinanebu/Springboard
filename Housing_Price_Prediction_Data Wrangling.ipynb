# Capstone Project: House Price Prediction


### Step 1: Importing Libraries and Loading the Dataset
We begin by importing the necessary libraries and loading the dataset.

```python
import pandas as pd

# Load the dataset

housing_data = pd.read_csv('Housing.csv')

# Display the first few rows of the dataset
housing_data.head()
```

### Step 2: Initial Inspection
Inspect the dataset for missing values, data types, and overall structure.

```python
# Check for missing values
missing_values = housing_data.isnull().sum()
print("Missing Values:")
print(missing_values)

# Get dataset info and summary statistics
housing_data.info()
housing_data.describe(include='all')
```

### Observations
- No missing values were found in the dataset.
- The dataset contains both numerical and categorical variables.
- Numerical variables like `price`, `area`, `bedrooms`, and others need to be analyzed for outliers.
- Categorical variables like `mainroad`, `guestroom`, `basement` need to be encoded.

### Step 3: Handling Categorical Variables
We encode binary categorical variables (`yes`/`no`) into `1`/`0` and apply one-hot encoding to multiclass categorical variables.

```python
# Encode binary categorical variables
binary_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
housing_data[binary_columns] = housing_data[binary_columns].replace({'yes': 1, 'no': 0})

# Apply one-hot encoding for 'furnishingstatus'
housing_data = pd.get_dummies(housing_data, columns=['furnishingstatus'], drop_first=True)

# Display updated dataset
housing_data.head()
```

### Step 4: Outlier Detection and Handling
We inspect numerical variables for potential outliers and take necessary steps.

```python
# Visual inspection of numerical features for outliers
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
for column in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=housing_data[column])
    plt.title(f"Boxplot for {column}")
    plt.show()
```

### Observations
- Potential outliers are observed in `price` and `area`. These will be handled as needed.

```python
# Remove extreme outliers in 'price' and 'area'
housing_data = housing_data[(housing_data['price'] < housing_data['price'].quantile(0.99)) &
                             (housing_data['area'] < housing_data['area'].quantile(0.99))]

# Confirm outliers removal
print("Dataset shape after removing outliers:", housing_data.shape)
```

### Step 5: Final Dataset Preparation
The dataset is now ready for further analysis or modeling. Below is the clean, preprocessed data.

```python
# Display final dataset structure
housing_data.head()
```

The dataset has been successfully cleaned and preprocessed, including:
- Handling missing values (none found).

- Detecting and handling outliers.
